```python
path_img = config.root + '/data/rs_img/pine_island_S3A_20180207T115228_20180207T115528.tif'
s3_img, s3_img_info = readTiff(path_img)
print(s3_img_info)
```

### 1. split into tiles and merged tiles.

```python
## show splited tiles
with h5py.File('data/icesat2/processed_atl06/pineisland_atl06_AD.h5','r') as f_a:
# with h5py.File('data/icesat2/processed_atl03/pineisland_atl03_spot1_A.h5','r') as f_a:
    lat_merge   = f_a['lat'][:]
    lon_merge   = f_a['lon'][:]
    h_elv_merge = f_a['h_elv'][:]
    t_yrs_merge = f_a['t_year'][:]

with h5py.File('data/icesat2/processed_atl06/pineisland_atl06_AD_buff_0_epsg_3031_tile_001.h5','r') as f_a:
# with h5py.File('data/icesat2/processed_atl03/pineisland_atl03_spot1_A_buff_0_epsg_3031_tile_001.h5','r') as f_a:
    lat_tile   = f_a['lat'][:]
    lon_tile   = f_a['lon'][:]
    h_elv_tile = f_a['h_elv'][:]
    t_yrs_tile = f_a['t_year'][:]

with h5py.File('data/icesat2/processed_atl06/pineisland_atl06_AD_buff_0_espg_3031_tiles_merge.h5','r') as f_a:
# with h5py.File('data/icesat2/processed_atl03/pineisland_atl03_spot1_A_buff_0_espg_3031_tiles_merge.h5','r') as f_a:
    lat_tiles_merge   = f_a['lat'][:]
    lon_tiles_merge   = f_a['lon'][:]
    h_elv_tiles_merge = f_a['h_elv'][:]
    t_yrs_tiles_merge = f_a['t_year'][:]

x_ad, y_ad = coor2coor(srs_from=4326, srs_to=s3_img_info['geosrs'], x=lon_merge, y=lat_merge)
x_tile, y_tile = coor2coor(srs_from=4326, srs_to=s3_img_info['geosrs'], x=lon_tile, y=lat_tile)
x_tiles_merge, y_tiles_merge = coor2coor(srs_from=4326, srs_to=s3_img_info['geosrs'], x=lon_tiles_merge, y=lat_tiles_merge)
```

```python
fig = plt.figure(figsize=(14,6))
plt.subplot(1,3,1)
imgShow(s3_img, extent=s3_img_info['geoextent'], color_bands=(0, 1, 2), clip_percent=5)
plt.scatter(x_ad[::100], y_ad[::100], s=3, c = h_elv_merge[::100], alpha=.7, cmap='terrain')
plt.clim([100,1400])
plt.title('atl06')
plt.colorbar(fraction=0.0320, pad=0.02, label='Elevation (m)')

plt.subplot(1,3,2)
imgShow(s3_img, extent=s3_img_info['geoextent'], color_bands=(0, 1, 2), clip_percent=5)
plt.scatter(x_tile[::100], y_tile[::100], s=3, c = h_elv_tile[::100], alpha=.7, cmap='terrain')
plt.clim([100,1400])
plt.yticks([])
plt.title('atl06_tile')
plt.colorbar(fraction=0.0320, pad=0.02, label='Elevation (m)')

plt.subplot(1,3,3)
imgShow(s3_img, extent=s3_img_info['geoextent'], color_bands=(0, 1, 2), clip_percent=5)
plt.scatter(x_tiles_merge[::100], y_tiles_merge[::100], s=3, c = h_elv_tiles_merge[::100], alpha=.7, cmap='terrain')
plt.clim([100,1400])
plt.yticks([])
plt.title('atl06_tiles_merge')
plt.colorbar(fraction=0.0320, pad=0.02, label='Elevation (m)')
```
